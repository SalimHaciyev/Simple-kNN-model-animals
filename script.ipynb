```python
import cv2
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from numpy import array
```

```python
def func(animal):
    lst = []
    for x in range(1, 1001):
        num = '{0:005}'.format(x)
        img_path = 'C:/Users/ASUS TUF/Desktop/Quick Access/ml/csv/animals pics/animals/' + animal + '/' + animal + '_' + num + '.jpg'
        img2 = cv2.imread(img_path, 0)
        img2 = cv2.resize(img2, (100, 100), interpolation = cv2.INTER_AREA)
        img2 = img2 / 255
        img2 = img2.reshape(1, -1)
        lst.append(img2)
    return lst

dogs = func('dogs')
cats = func('cats')
panda = func('panda')
```

```python
x = [dogs, cats, panda]
x = array(x).reshape(3000, -1)

y = []
for i in range(3):
    for j in range(1000):
        y.append(i)
```

```python
print(array(y).shape)
print(x.shape)
```

```python
classifier = KNeighborsClassifier(n_neighbors = 3)
classifier.fit(x, y)
```

```python
def Model(name):
    path ='C:/Users/ASUS TUF/Desktop/Quick Access/dl/Pics for models/animals/' + str(name )+ '.jpg'
    a = cv2.imread(path, 0)
    a = cv2.resize(a, (100, 100), interpolation = cv2.INTER_NEAREST)
    plt.imshow(a)
    a = a / 255
    a = a.reshape(1, -1)
    result = classifier.predict(a)
    
    if result == 0:  print('Dog')
    elif result == 1:  print('Cat')
    else:  print('Panda')
```
